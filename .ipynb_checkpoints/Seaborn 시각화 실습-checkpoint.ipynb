# 2. 타이타닉 데이터를 통한 시각화 실습

## (1) 데이터 준비

```python
# 데이터 출처 : https://www.kaggle.com/c/titanic

train = pd.read_csv("data/train.csv")
train.head()
```

```python
train.shape
train.info()
```

```python
# 결측치 확인
train.isnull().sum()
```

## (2) EDA(Exploratory Data Analysis)

```python
def bar_chart(feature):
    survived = train[train["Survived"]==1][feature].value_counts()
    dead = train[train["Survived"]==0][feature].value_counts()

    df = pd.DataFrame([survived, dead])
    df.index = ["Survived", "Dead"]
    df.plot(kind="bar", stacked=True, figsize=(6, 3))
```

##### 성별에 따른 생존 여부

```python
bar_chart("Sex")
```

##### 객실 등급(Pclass)에 따라 생존여부 관계확인

```python
bar_chart("Pclass")
```

##### 형제자매, 배우자 동승여부(SibSp)에 따라 생존여부 관계확인

```python
bar_chart("SibSp")
```

##### 부모나 자식 동승여부(Parch)에 따라 생존여부 관계확인

```python
bar_chart("Parch")
```

##### 경유지(Embarked)에 따라 생존여부 관계확인

```python
bar_chart("Embarked")
```
